```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['axes.unicode_minus'] = False # - 부호 깨짐 방지
%matplotlib inline
```

```python
# 한글 폰트 문제 해결
import platform

from matplotlib import font_manager, rc
# plt.rcParams['axes.unicode_minus'] = False

if platform.system() == 'Darwin':
    rc('font', family='AppleGothic')
elif platform.system() == 'Windows':
    path = "c:/Windows/Fonts/malgun.ttf"
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print('Unknown system... sorry~~~~') 
```

```python
# pie chart : 카테고리별 값의 상대적 비교
# 차트 원형을 유지 : plt.axis('equal')

labels = ['뽕사부', '순대국밥', '비빔밥', '생고기', '서브웨이']
sizes = [15, 15, 20, 25, 25]
colors = ['gold', 'blue', 'red', 'yellow', 'green']

plt.title('Pie Chart')
plt.pie(sizes, labels = labels, colors = colors, shadow = True, autopct = '%1.1f%%')
plt.axis('equal')
plt.show()
```

```python
# histogram : 구간에 대한 데이터 집계 시각화

data = np.random.randn(1000)

plt.title('histogram')
plt.hist(data, bins = 10) # bins 가 늘어날수록 밀집된다.
plt.show()
```

```python
# scatter plot : 2차원 데이터
# 데이터 집합의 상관관계를 살펴보기 위한 시각화

np.random.seed(100)
X = np.random.normal(0, 1, 100)
Y = np.random.normal(0, 1, 100)

plt.title('scatter plot')
plt.scatter(X, Y)
plt.show()
```

```python
np.random.seed(100)
X = np.random.rand(30)
Y1 = np.random.rand(30)
Y2 = np.random.rand(30)
Y3 = np.pi * (15 * np.random.rand(30)) ** 2

plt.title('bubble plot')
plt.scatter(X, Y1, c = Y2, s = Y3)
plt.show()
```

```python
import seaborn as sns
iris = sns.load_dataset('iris', engine = 'python')
titanic = sns.load_dataset('titanic', engine = 'python')
```

```python
iris.sepal_length[:20].plot(kind = 'bar', rot = 0)
plt.show()
```

```python
# iris[:5].plot(kind = 'bar', rot = 0)
iris[:5].plot.bar(rot = 0)
plt.show()
```

```python
# 그룹분석으로 각 붓꽃의 평균을 구하고 시각화
iris.groupby('species').mean().plot(kind = 'bar', rot = 0)
plt.grid()
plt.title('붓꽃 종별 데이터 평균')
plt.xlabel('종')
plt.ylabel('평균')
plt.yticks([0, 2, 4, 6, 8])
plt.show()
```

```python
# 선실별 승객 수 비율을 파이차트로 구현
titanic.head()

temp = titanic['pclass'].value_counts()
temp.plot.pie(autopct = '%.2f%%', startangle = 45, legend = True)
plt.axis('equal')
plt.show()
```

#### autopct 사용법
- '%0.2f' — 소숫점 두번째자리 까지 보여준다. 4.08
- '%0.2f%%' — 소숫점 두번째자리까지 보여주고 맨 뒤에 ‘%’ 기호를 붙인다. 4.08%.
- (% 기호를 두 번 쓰는 이유는 문자열 포매팅 할 때 처음 % 기호가 이스케이프 문자처럼 작동하기 때문이다.)
- '%d%%' — 가장 가까운 정수로 반올림하고 맨 뒤에 % 기호를 붙인다.
