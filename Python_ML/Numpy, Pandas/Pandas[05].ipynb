```python
iris = sns.load_dataset('iris', engine = 'python')
iris.head()
```

```python
iris.describe()
```

```python
# 각 종별로 가장 큰 값과 가장 작은 값의 비율 구하기

# aggregate : agg()
# 우리가 원하는 집계함수가 없으면 만들어서 쓴다.
def get_ratio(x) :
    return x.max() / x.min()
```

```python
iris.groupby(iris.species).sum()
```

```python
iris.groupby(iris.species).mean()
```

```python
iris.groupby(iris.species).agg(get_ratio)
```

```python
iris.groupby(iris.species).describe().T
```

```python
# 붓꽃의 종별 가장 큰 꽃잎길이 3개의 데이터를 추출하는 함수 정의
def max3_petal_length_func(df) :
    return df.sort_values(by = 'petal_length', ascending = False)[:3]
```

```python
iris.groupby(iris.species).apply(max3_petal_length_func)
```

```python
# transform
# 데이터 프레임 자체를 변화시키는 함수
# 원본 프레임과 크기가 같다.

# cut() : 동일 길이로 나누어서 범주를 만들어서 그룹에 대한 통계량을 구한다.
# qcut() : 동일 개수로 나누어서 범주를 만들어서 그룹에 대한 통계량을 구한다.

# 각 붓꽃 길이가 해당 종 내에서 대, 중, 소 어느 것에 해당되는지에 대한 데이터프레임 만들기
iris.groupby(iris.species).groups
```

```python
iris.groupby(iris.species)
```

```python
iris.groupby(iris.species).petal_length
```

```python
iris.groupby(iris.species).petal_length.size()
```

```python
def cat3_petal_length(s) :
    return pd.qcut(s, 3, labels = ['소', '중', '대']).astype(str)
```

```python
iris['category'] = iris.groupby(iris.species).petal_length.transform(cat3_petal_length)
iris.head()
```

```python
# 붓꽃 데이터에서 붓꽃 종별로 꽃잎 길이, 꽃잎 폭 등의 평균 구하기
display(iris.groupby(iris.species).petal_length.mean())
print('*' * 50)

display(iris.groupby(iris.species).sepal_width.mean())
print('*' * 50)

# agg : 여러개의 집계함수를 넣어서 실행 가능
# apply : 하나의 함수만 실행 가능
iris.groupby(iris.species).agg(np.mean).loc[:, ['sepal_length', 'sepal_width']]
```

### pivot
- 데이터프레임에서 두 개의 열을 이용하여 행/열 인덱스 reshape 된 테이블을 의미한다.
- 새로운 테이블에서 새로운 기준으로 집계
- pivot(index, columns, values)
- pivot_table(data, values, index, columns, aggfunc = 'mean')

```python
titanic = sns.load_dataset('titanic', engine = 'python')
titanic.head()
```

```python
# pivot()
titanic_df01 = pd.DataFrame(titanic, columns = ['sex', 'pclass'])
titanic_df01.head()
```

```python
titanic_df01 = titanic.groupby(['sex', 'pclass']).size().reset_index(name = 'cnt')
titanic_df01
```

```python
titanic_df01.pivot('sex', 'pclass', 'cnt')
```

```python
# 성별과 생존여부에 따른 승객 수 집계
titanic_df02 = titanic.groupby(['sex', 'survived']).size().reset_index(name = 'cnt')
display(titanic_df02)
print('*' * 50)

titanic_df02.pivot('sex', 'survived', 'cnt')
```

```python
data = {
    "도시": ["서울", "서울", "서울", "부산", "부산", "부산", "인천", "인천"],
    "연도": ["2015", "2010", "2005", "2015", "2010", "2005", "2015", "2010"],
    "인구": [9904312, 9631482, 9762546, 3448737, 3393191, 3512547, 2890451, 263203],
    "지역": ["수도권", "수도권", "수도권", "경상권", "경상권", "경상권", "수도권", "수도권"]
}
columns = ["도시", "연도", "인구", "지역"]

pivot_sample_df = pd.DataFrame(data, columns=columns)
pivot_sample_df
```

```python
pivot_sample_df.pivot_table('인구', '도시', '연도')
```

```python
# pivot_table 옵션
# data : 데이터프레임
# values : 분석할 열
# index : 행
# columns : 열
# aggfunc : 집계함수
# fill_value : NaN 대체값
# margins : 분석 결과를 오른쪽과 아래에 붙일지 여부
# margins_name : margin 열(행)의 이름

pivot_sample_df.pivot_table('인구', '도시', '연도', margins = True, margins_name = '분석결과')
```

```python
titanic = sns.load_dataset('titanic', engine = 'python')
titanic.head()
```

```python
# 1. qcut 명령으로 세 개의 나이 그룹을 만든다.
# 0~19  : A
# 20~59 : B
# 60 ~  : C

titanic['age_cut'] = pd.qcut(titanic['age'], 3, labels = ['A', 'B', 'C'])
titanic.head()
```

```python
# 2. 성별, 선실, 나이 그룹에 의한 생존율을 데이터프레임으로 계산한다. 
# 행에는 성별 및 나이 그룹에 대한 다중 인덱스를 사용하고
# 열에는 선실 인덱스를 사용한다. 
# 생존율은 해당 그룹의 생존 인원수를 전체 인원수로 나눈 값이다.

count_df = titanic.pivot_table('survived', ['sex', 'age_cut'], 'pclass', aggfunc = 'count', margins = True, margins_name = 'sur_ratio')
all_count = count_df.loc['sur_ratio', 'sur_ratio'][0]
survival_rate_df = count_df / all_count

display(count_df)
print('*' * 50)

display(survival_rate_df)
```

```python
# 3. 성별 및 선실에 의한 생존율을 pivot 데이터 형태로 만든다.
count_df = titanic.pivot_table('survived', 'sex', 'pclass', aggfunc = 'sum', margins = True, margins_name = 'sur_ratio')
all_count = titanic.shape[0]
survival_rate_df = count_df / all_count

print(all_count)
survival_rate_df
```
