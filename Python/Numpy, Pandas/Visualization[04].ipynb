```python
import seaborn as sns
```

```python
iris = sns.load_dataset('iris', engine = 'python')
titanic = sns.load_dataset('titanic', engine = 'python')
tips = sns.load_dataset('tips', engine = 'python')
flights = sns.load_dataset('flights', engine = 'python')
```

```python
# bar plot : 카테고리 값에 따른 실수의 값의 평균과 편차를 표시
# 평균 : 막대의 높이, 편차 : error bar

# 스타일 테마 설정 (darkgrid, whitegrid, dart, white, ticks)
sns.set_style('darkgrid')

# 그래프 객체를 생성하기 위한 figure 객체 생성
fig = plt.figure(figsize = (15, 5))

area01 = fig.add_subplot(1, 3, 1)
area01.set_title('titanic survived - sex')

area02 = fig.add_subplot(1, 3, 2)
area02.set_title('titanic survived - sex hue')

area03 = fig.add_subplot(1, 3, 3)
area03.set_title('titanic survived - sex hue dodge')

# 성별에 따른 생존율
sns.barplot(x = 'sex', y = 'survived', data = titanic, ax = area01)

# hue
sns.barplot(x = 'sex', y = 'survived', hue = 'class', data = titanic, ax = area02)

# dodge
sns.barplot(x = 'sex', y = 'survived', hue = 'class', dodge = False, data = titanic, ax = area03)

plt.show()
```

```python
tips.head()
```

```python
# tips bar plot
# 요일별 전체 팁 시각화
sns.barplot(x = 'day', y = 'total_bill', data = tips)
plt.title('요일별 전체 팁')
plt.show()
```

```python
# count plot() : 카테고리 값별 데이터 빈도를 표시
# x축의 빈도를 세는 것. y축에 데이터를 명시하지 않는다.

# 스타일 테마 설정 (darkgrid, whitegrid, dart, white, ticks)
sns.set_style('whitegrid')

# 그래프 객체를 생성하기 위한 figure 객체 생성
fig = plt.figure(figsize = (15, 5))

area01 = fig.add_subplot(1, 3, 1)
area01.set_title('titanic class')

area02 = fig.add_subplot(1, 3, 2)
area02.set_title('titanic class - who')

area03 = fig.add_subplot(1, 3, 3)
area03.set_title('titanic class - who(stacked)')

# 기본
sns.countplot(x = 'class', data = titanic, ax = area01)

# hue
sns.countplot(x = 'class', hue = 'who', palette = 'Set2', data = titanic, ax = area02)

# dodge
sns.countplot(x = 'class', hue = 'who', dodge = False, palette = 'Set3', data = titanic, ax = area03)

plt.show()
```

```python
# tips count plot()
# 요일별로 팁을 준 빈도 시각화
sns.countplot(x = 'day', data = tips)
plt.title('요일별 팁을 준 횟수')
plt.show()
```

```python
# stripplot() : 스카터플롯처럼 모든 데이터를 점으로 시각화
# jitter = True : 위치를 무작위로 바꿔서 겹치지 않게 해준다.
```

```python
sns.stripplot(x = 'day', y = 'total_bill', jitter = True, data = tips)
plt.title('요일별 팁을 준 횟수')
plt.show()
```

```python
sns.swarmplot(x = 'day', y = 'total_bill', data = tips)
plt.title('요일별 팁을 준 횟수')
plt.show()
```

```python
# titanic
# 서브플롯을 2개 영역으로 만들고
# class 별 나이의 분포 확인 (stripplot(), swarmplot())

# 그래프 객체를 생성하기 위한 figure 객체 생성
fig = plt.figure(figsize = (15, 5))

area01 = fig.add_subplot(1, 2, 1)
area01.set_title('titanic class - strip')

area02 = fig.add_subplot(1, 2, 2)
area02.set_title('titanic class - swarm')

sns.stripplot(x = 'class', y = 'age', jitter = True, data = titanic, ax = area01)
sns.swarmplot(x = 'class', y = 'age', data = titanic, ax = area02)

plt.show()
```

```python
# tips
# 요일별, 성별 전체 팁 stripplot()시각화
sns.stripplot(x = 'day', y = 'total_bill', hue = 'sex', data = tips, jitter = True )
plt.title('요일별 전체 팁')
plt.legend(loc = 'best')
plt.show()
```

```python
# 스타일 테마 설정 (darkgrid, whitegrid, dart, white, ticks)
sns.set_style('whitegrid')

# 그래프 객체를 생성하기 위한 figure 객체 생성
fig = plt.figure(figsize = (15, 10))

ax01 = fig.add_subplot(2, 2, 1)
ax01.set_title('box')

ax02 = fig.add_subplot(2, 2, 2)
ax02.set_title('box')

ax03 = fig.add_subplot(2, 2, 3)
ax03.set_title('violin')

ax04 = fig.add_subplot(2, 2, 4)
ax04.set_title('violin')

sns.boxplot(x = 'alive', y = 'age', data = titanic, ax = ax01)
sns.boxplot(x = 'alive', y = 'age', data = titanic, hue = 'sex', ax = ax02)

sns.violinplot(x = 'alive', y = 'age', data = titanic, ax = ax03)
sns.violinplot(x = 'alive', y = 'age', data = titanic, hue = 'sex', ax = ax04)

plt.show()
```

```python
# heatmap : 하나 이상의 카테고리 값의 변화를 색깔 변화로 시각화
# pivot 테이블로 범주형 변수를 각각 행, 열로 재구분하여 정리
# 인덱스 : 성별, 열 : 클래스

table = titanic.pivot_table(index = ['sex'], columns = ['class'], aggfunc = 'size')
table.head()

sns.heatmap(table, annot = True, fmt = 'd', cbar = False, linewidth = 1)
plt.show()
```
