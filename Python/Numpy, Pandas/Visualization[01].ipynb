```python
Y = [2, 3, 1]
X = np.arange(len(Y))

xlabel = ['1등실', '2등실', '3등실']

plt.title('bar plot')
plt.bar(X, Y)
plt.xticks(X, xlabel)
plt.yticks(sorted(Y))
plt.xlabel('선실 등급')
plt.ylabel('생존자 수')
plt.show()
```

```python
Y = [2, 3, 1]
X = np.arange(len(Y))

xlabel = ['1등실', '2등실', '3등실']

plt.title('bar plot')
plt.barh(X, Y)
plt.xticks(sorted(Y))
plt.yticks(X, xlabel)
plt.xlabel('생존자 수')
plt.ylabel('선실 등급')
plt.show()
```

```python
# 양적 자료의 데이터 분포 확인 (boxplot)
data_df.boxplot(['displ', 'cyl', 'cty', 'hwy'])
plt.show()
```

```python
data_df['class'].value_counts()
```

```python
# 데이터프레임 query 함수
query_df = pd.DataFrame(np.random.randint(0, 10, size = (10, 3)), columns = ['col01', 'col02', 'col03'])
query_df
```

```python
query_df.query('col01 > 2 and col02 < 5')
```

```python
# 자동차 배기량(displ)에 따라 고속도로 연비(hwy)의 변화
# 배기량이 4 이하인 자동차와 5 이상인 자동차 중 어떤 자동차의 고속도로연비가 평균적으로 높은지 확인

print(data_df[data_df['displ'] <= 4]['hwy'].mean())
print(data_df[data_df['displ'] >= 5]['hwy'].mean())
```

```python
# 자동차 제조 회사에 따라 도시 연비 변화
# audi 와 toyota 중 어느 manufacturer(제조사)의 cty(도시연비)가 평균적으로 더 높은지 확인

print(data_df.query('manufacturer == "audi"')['cty'].mean())
print(data_df.query('manufacturer == "toyota"')['cty'].mean())

# data_df[['manufacturer', 'cty']].boxplot(by = 'manufacturer')
```

```python
# chevrolet, ford, honda 자동차의 고속도로 연비 평균 확인
# 이 회사들의 데이터를 추출한 후 hwy 전체 평균을 확인

chevrolet = data_df[data_df['manufacturer'] == 'chevrolet']
ford = data_df[data_df['manufacturer'] == 'ford']
honda = data_df[data_df['manufacturer'] == 'honda']

print(chevrolet['hwy'].mean())
print(ford['hwy'].mean())
print(honda['hwy'].mean())
print()

# 1
data_df02 = data_df[(data_df['manufacturer'] == 'chevrolet') | (data_df['manufacturer'] == 'ford') | (data_df['manufacturer'] == 'honda')]
print(data_df02.hwy.mean())
print('*' * 50)

# 2
manufac_car = ['chevrolet', 'ford', 'honda']
print(data_df.query('manufacturer in @manufac_car')['hwy'].mean()) # @ : 외부 변수를 가져다 쓸 경우 사용
```

```python
chevrolet_index = data_df['manufacturer'] == 'chevrolet'
ford_index = data_df['manufacturer'] == 'ford'
honda_index = data_df['manufacturer'] == 'honda'

temp_df = data_df.loc[chevrolet_index | ford_index | honda_index, :]
mean_df = temp_df[['manufacturer', 'hwy']].groupby(temp_df.manufacturer).agg(np.mean)
mean_df.plot.bar(rot = 0)

display(mean_df)
```

```python
data_df.head()
```

```python
# 데이터 전처리 : 연료가격 변수 추가 (price_fl)
# c : 가스 2.35, d : 디젤 2.38, e : 에탄올 2.11, p : 고급휘발유 2.76, r : 보통휘발유 2.22

fl_df = pd.DataFrame({
    'fl' : ['c', 'd', 'e', 'p', 'r'],
    'price_fl' : [2.35, 2.38, 2.11, 2.76, 2.22]
})

fl_df
```

```python
# 연료 가격 프레임을 원본 프레임에 병합
data_fl_merge_df = pd.merge(data_df, fl_df, on = 'fl')
data_fl_merge_df.head()
```

```python
# merge한 데이터프레임에서 model, fl, price_fl 추출
data_fl_merge_df[['model', 'fl', 'price_fl']].head()
```

```python
# 데이터 전처리과정에서 결측치 개수 확인
data_fl_merge_df.isna().sum()
```

```python
# 구동방식(drv) 별 고속도로 연비(hwy) 평균
# 임의적으로 결측값 처리를 위해서 더미 값 넣기
data_fl_merge_df.loc[65, 'hwy'] = np.nan
data_fl_merge_df.loc[120, 'hwy'] = np.nan
data_fl_merge_df.loc[154, 'hwy'] = np.nan
data_fl_merge_df.loc[189, 'hwy'] = np.nan
data_fl_merge_df.loc[219, 'hwy'] = np.nan
data_fl_merge_df.loc[230, 'hwy'] = np.nan
```

```python
data_fl_merge_df.isna().sum()
```

```python
data_fl_merge_df[['drv', 'hwy']].isna().sum()
```

```python
data_fl_merge_df['drv'].value_counts()
```

```python
# hwy 변수의 결측값을 제외하고 어떤 구동방식의 고속도로평균 연비가 높은지 확인
data_fl_merge_isna_df = data_fl_merge_df[['drv', 'hwy']].dropna()
data_fl_merge_isna_df.info()

data_fl_merge_isna_df.groupby('drv').mean()
```

```python
# 구동 방식을 연비평균을 비교하기 위한 막대그래프로 시각화
temp_df = data_fl_merge_isna_df.groupby('drv').mean()
temp_df

temp_df.plot.bar(rot = 0)
plt.title('구동방식별 평균연비')
plt.xlabel('구동방식')
plt.ylabel('연비')
plt.grid()
plt.show()
```

```python
# 구동 방식별 고속도로, 도시연비의 평균 계산
# 데이터프레임으로 만들어서 bar plot 시각화

temp_df = data_df.groupby('drv')[['cty', 'hwy']].mean()

temp_df.plot.bar(rot = 0)
plt.title('구동 방식별 고속도로/도시 연비')
plt.grid()
plt.xlabel('구동 방식')
plt.ylabel('연비')
```

```python
# 자동차(class) 중에서 어떤 자동차가 가장 많은지 확인
# 종류별 빈도를 bar plot 시각화

temp_df = data_df.groupby('class')[['fl']].count()
temp_df.columns = ['count']

temp_df.plot.bar(rot = 0)
plt.title('자동차 종류별 대수')
plt.grid()
plt.xlabel('차종')
plt.ylabel('대수')
```

```python
# 어떤 회사에서 생산한 suv 차종의 도시 연비가 높은지 확인
# suv 차종을 대상으로 평균 도시연비가 가장 높은 회사 다섯곳을 bar plot 시각화

temp_df = data_df[data_df['class'] == 'suv'].groupby('manufacturer')[['cty']].mean()
temp_df = temp_df.sort_values('cty', ascending = False).head()

temp_df.plot.bar(rot = 0)
plt.title('제조사별 suv에 대한 평균 도시연비')
plt.grid()
plt.xlabel('제조사')
plt.ylabel('도시연비')
```
