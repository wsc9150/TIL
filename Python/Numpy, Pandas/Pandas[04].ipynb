```python
import seaborn as sns

titanic = sns.load_dataset('titanic', engine = 'python')
```

```python
# 타이타닉 승객 중 나이를 명시하지 않은 고객은 나이를 명시한 고객의 평균 나이 값으로 대체

# 타이타닉 승객에 대해 나이와 성별에 의한 age_gender_cat 열을 만들고
# 조건1. 성별을 나타내는 문자열 male 또는 female로 시작한다.
# 조건2. 성별을 나타내는 문자열 뒤에 나이를 나타내는 문자열로 변경
# 조건3. 예시) 남성의 나이가 27이라면 -> male27

# display(titanic.head(10))

avg_age = round(titanic['age'].mean())
# print(avg_age)

titanic['age'] = titanic['age'].fillna(avg_age)
# display(titanic.head(10))

titanic['age_gender_cat'] = titanic['sex'] + titanic['age'].astype(int).astype(str)
display(titanic.head(10))

# print(type(titanic['age'][0]))
```

```python
# 5명의 학생의 국어, 영어, 수학 점수를 나타내는 데이터프레임을 만든다.

# 1. 
# 학생 이름을 나타내는 열을 포함시키지 않고 
# 데이터프레임 df_score1 을 생성한 후, 
# df_score1.index 속성에 학생 이름을 나타내는 열을 지정하여 인덱스를 지정한다. 
# reset_index 명령으로 이 인덱스 열을 명령으로 
# 일반 데이터열로 바꾸여 데이터프레임 df_score2을 만든다.
df_score1 = pd.DataFrame({
    'kor' : [80, 90, 50, 40, 70],
    'eng' : [100, 90, 60, 70, 80],
    'math' : [90, 50, 70, 80, 80]
})

df_score1

df_score1.index = ['A', 'B', 'C', 'D', 'E']
df_score1.index.name = '이름'

df_score2 = df_score1.reset_index()
display(df_score2)

# 2.
# 학생 이름을 나타내는 열이 일반 데이터 열을 포함하는 데이터프레임 df_score2에 
# set_index 명령을 적용하여 다시 학생 이름을 나타내는 열을 인덱스로 변경한다.
df_score2 = df_score2.set_index('이름')
display(df_score2)
```

```python
# DataFrame merge
data1 = {
    '학번' : [1, 2, 3, 4],
    '이름' : ['섭섭해', '김한준', '김선림', '최호진'],
    '학년' : [2, 4, 1, 3]
}

data2 = {
    '학번' : [1, 2, 4, 5],
    '학과' : ['CS', 'MATH', 'MATH', 'CS'],
    '학점' : [2.4, 4.5, 3.4, 4.2]
}
```

```python
stu_df = pd.DataFrame(data1)
major_df = pd.DataFrame(data2)
display(stu_df)
display(major_df)
```

```python
pd.merge(stu_df, major_df, on = '학번', how = 'inner') # default 옵션.
```

```python
pd.merge(stu_df, major_df, on = '학번', how = 'outer')
```

```python
# DataFrame merge
data1 = {
    '학번' : [1, 2, 3, 4],
    '이름' : ['섭섭해', '김한준', '김선림', '최호진'],
    '학년' : [2, 4, 1, 3]
}

data2 = {
    '과목코드' : [1, 2, 4, 5],
    '학과' : ['CS', 'MATH', 'MATH', 'CS'],
    '학점' : [2.4, 4.5, 3.4, 4.2]
}
```

```python
stu_df = pd.DataFrame(data1)
major_df = pd.DataFrame(data2)
display(stu_df)
display(major_df)
```

```python
# 공통된 column이 없는 경우 merge
pd.merge(stu_df, major_df, left_on = '학번', right_on = '과목코드', how = 'inner')
```

```python
# column 인덱스가 아닌 인덱스를 기준으로 병합
# left_index, right_index

pop_df01 = pd.DataFrame({
    'city' : ['seoul', 'seoul', 'seoul', 'busan', 'busan'],
    'year' : [2010, 2005, 2020, 2018, 2015],
    'pop' : [1234567, 2345678, 3456789, 4567890, 5678901]
})

pop_df01
```

```python
# 다중 인덱스
pop_df02 = pd.DataFrame(np.arange(12).reshape((6, 2)),
                        index = [['busan', 'busan', 'seoul', 'seoul', 'seoul', 'seoul'], [2010, 2005, 2020, 2018, 2015, 2010]],
                        columns = ['col1', 'col2']
)

pop_df02
```

```python
pd.merge(pop_df01, pop_df02, left_on = ['city', 'year'], right_index = True)
```
