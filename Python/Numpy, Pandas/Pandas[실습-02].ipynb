```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# [실습]
# ratings.dat
# movies.dat
# users.dat

# 1. 각 파일을 pd.read_csv() 로 읽어들인다
# - 헤더 및 구분자 확인요망
ratings_df = pd.read_csv('./data/ratings.dat', delimiter = '::', header = None)
users_df = pd.read_csv('./data/users.dat', delimiter = '::', header = None)
movies_df = pd.read_csv('./data/movies.dat', delimiter = '::', header = None)

display(ratings_df.head())
display(movies_df.head())
display(users_df.head())
```

```python
# 2. 데이터 확인 및 구조분석과 각 데이터 프레임에 헤더를 추가
# ratings.dat
# - userId movieId rating timestamp
ratings_df.columns = ['userId', 'movieId', 'rating', 'timestamp']

# movies.data
# - movieId title genres
movies_df.columns = ['movieId', 'title', 'genres']

# users.data
# 뒤에 두 열은 제거
# - userId gender age
users_df.drop([3, 4], axis = 1, inplace = True)
users_df.columns = ['userId', 'gender', 'age']
```

```python
display(ratings_df.head())
display(movies_df.head())
display(users_df.head())
```

```python
# 3. 공통 컬럼을 이용한 데이터 병합
merge_df01 = pd.merge(ratings_df, users_df, on = 'userId')
merge_df01.head()

merge_df02 = pd.merge(merge_df01, movies_df, on = 'movieId')
merge_df02.head()
```

```python
# 4. user 평점 갯수 순위 확인
user_sort = pd.DataFrame(merge_df02.groupby('userId').size())
user_sort.head()

user_sort.sort_values(by = 0, ascending = False).head(10)
```

```python
# 5. 각 영화의 성별 평균 평점
# groupby
display(merge_df02.groupby(['title', 'gender'])[['rating']].mean().head(20))
print('*' * 50)

# pivot_table
display(merge_df02.pivot_table('rating', 'title', 'gender', aggfunc = 'mean').head(10))
```

```python
# 6. 평점이 220개 이상인 영화만 필터링
rating_by_title = merge_df02.groupby('title').size()
rating220 = rating_by_title.index[rating_by_title >= 220]
rating220
```

```python
# 7. 평점이 200개 이상인 영화들의 성별 평균 평점
s = merge_df02['movieId'].value_counts() >= 200
movie_list = []
male_avg = []
female_avg = []
title_list = []

for key, value in s.items() :
    if value :
        movie_list.append(key)
        
for movie in movie_list :
    over200 = merge_df02[merge_df02['movieId'] == movie]
    
    male = over200[over200['gender'] == 'M']
    male_avg.append(round(male['rating'].mean(), 1))
    
    female = over200[over200['gender'] == 'F']
    female_avg.append(round(female['rating'].mean(), 1))
    
    title = movies_df[movies_df['movieId'] == movie]['title'].values
    title_list.extend(title)

result = pd.DataFrame({
    'title' : title_list,
    'male_avg' : male_avg,
    'female_avg' : female_avg
})

result.head()
```

```python
# 8. 여성이 가장 좋아하는 영화 순위
title_list = []
rating_list = []
female_like_df = merge_df02[merge_df02['gender'] == 'F']

movie_list = female_like_df['movieId'].unique()
# len(movie_list)

for movie in movie_list :
    rating_sum = female_like_df[female_like_df['movieId'] == movie]['rating'].sum()
    rating_list.append(rating_sum)
    
    title = movies_df[movies_df['movieId'] == movie]['title'].values
    title_list.extend(title)

rank_df = pd.DataFrame({
    'title' : title_list,
    'sum' : rating_list
})

rank_df = rank_df.sort_values(by = 'sum', ascending = False)
rank_df['rank'] = rank_df['sum'].rank(ascending = False).astype(int)
rank_df = rank_df.set_index('rank')

rank_df.head()
```

```python
# 9. 남녀간의 평균 평점 차이
male_df = merge_df02[merge_df02['gender'] == 'M']
male_avg = round(male_df['rating'].mean(), 2)

female_df = merge_df02[merge_df02['gender'] == 'F']
female_avg = round(female_df['rating'].mean(), 2)

round(abs(male_avg - female_avg), 2)
```

```python
# 10. 평점의 표준편차가 큰 영화
merge_df02.groupby('title')[['rating']].std()

std_sort = merge_df02.groupby('title')[['rating']].std().sort_values('rating', ascending = False)
std_sort.head()
```

```python
# 추가1. Comedy 영화 중 가장 평점이 낮은 영화의 제목
comedy_df = merge_df02[merge_df02['genres'] == 'Comedy']
movie_list = comedy_df['movieId'].unique()

min = 5
movie_id = ''

for movie in movie_list :
    rating_avg = comedy_df[comedy_df['movieId'] == movie]['rating'].mean()
    
    if min >= rating_avg :
        min = rating_avg
        movie_id = movie

movies_df[movies_df['movieId'] == movie_id][['title']]
```

```python
# 추가2. 평균 평점이 가장 높은 영화의 제목 (동률이 있을 경우 모두 출력)
movie_list = merge_df02['movieId'].unique()
avg_list = []

for movie in movie_list :
    avg = round(merge_df02[merge_df02['movieId'] == movie]['rating'].mean(), 4)
    avg_list.append(avg)

test_df = pd.DataFrame({
    'movieId' : movie_list,
    'avg' : avg_list
})


test_df = test_df.sort_values(by = 'avg', ascending = False)

movie_ids = test_df[test_df['avg'] == 5]['movieId']

movie_title = []
movie_id = []

for key, value in movie_ids.items() :
    movie_id.append(value)
    movie_title.extend(movies_df[movies_df['movieId'] == value]['title'].values)

result = pd.DataFrame({
    'movieId' : movie_ids,
    'Title' : movie_title,
})

result = result.reset_index(drop = True)
result
```
