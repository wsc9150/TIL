```python
# 파일 읽기
seoul_school_df = pd.read_excel('./data/elem_school_folium.xlsx')
```

```python
# Feature Selection
# 학교명, 학교급구분, 소재지도로명, 위도, 경도
feature_sel_df = seoul_school_df.filter(['학교명', '학교급구분', '소재지도로명주소', '위도', '경도'])
feature_sel_df.head()
```

```python
# 결측값 확인/제거
print(feature_sel_df.isna().sum())
print('*' * 50)

feature_sel_df.dropna(inplace = True)
print(feature_sel_df.isna().sum())
```

```python
# 소재지도로명주소가 서울을 포함하고 학교급구분이 초등학교인 데이터 추출
condition_df = feature_sel_df[(feature_sel_df['학교급구분'] == '초등학교') & (feature_sel_df['소재지도로명주소'].str.contains('서울특별시'))]
condition_df.head()
```

```python
# 지도 시각화
seoul_map = g.Map(location = [37.55, 126.98], zoom_start = 12)

for i in range(len(condition_df)) :
    marker = g.Marker(condition_df.iloc[i, [3, 4]], popup = condition_df.iloc[i]['학교명'], icon = g.Icon(color = 'red'))
    marker.add_to(seoul_map)

seoul_map
```
